```python
from edin.ccg.representation.derivation import *
from edin.ccg.representation.semantics import SemanticsAssigner

sem_assigner = SemanticsAssigner("English")
```

```python
tree_str = "(<T S 0 2> (<T S/NP 0 2> (<T S/NP 0 2> (<T S/(S\\NP) 0 1> (<T NP 0 2> (<L NP/N X X Every NP/N>) (<L N X X man N>) ) ) (<L (S\\NP)/NP X X loves (S\\NP)/NP>) ) (<T S/NP[conj] 0 2> (<L conj X X and conj>) (<T S/NP 0 2> (<T S/(S\\NP) 0 1> (<T NP 0 2> (<L NP/N X X every NP/N>) (<L N X X woman N>) ) ) (<L (S\\NP)/NP X X hates (S\\NP)/NP>) ) ) ) (<T NP 0 2> (<T NP 0 2> (<L NP/N X X some NP/N>) (<L N X X apple N>) ) (<T NP[conj] 0 2> (<L conj X X and conj>) (<T NP 0 2> (<L NP/N X X some NP/N>) (<L N X X orange N>) ) ) ) )"
tree = DerivationLoader.from_str(tree_str)
tree = sem_assigner.assign_semantics(tree)
tree.show_proof()
for r in tree.semantics.unpack_all_readings():
    r.show()
```
